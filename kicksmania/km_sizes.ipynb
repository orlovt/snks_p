```python
code = requests.get(url).text
```

```python
soup = BeautifulSoup(code, "html.parser")
```

```python
soup.select("div", {"class":"product__wrap clearfix"})
```

```python
section = soup.find(class_ = "product__wrap clearfix")
```

```python
section.find("h1").text
```

```python
section.find(class_="price").text
```

```python
def convert_price(sprice): 
    if "–" in sprice: 
        
        l = sprice.split("–")
        lprice = []
        for i in l: 
            lprice += int(i)
        print(lprice)

convert_price(section.find(class_="price").text)
```
